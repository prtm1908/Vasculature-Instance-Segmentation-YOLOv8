```python
import json
```

```python
with open('/kaggle/input/hubmap-hacking-the-human-vasculature/polygons.jsonl','r') as json_file:
    json_samples=[json.loads(line) for line in json_file]
```

```python
for i in json_samples:
    txtname=i['id']
    directory_path = '/kaggle/working/labels/train/'
    path = os.path.join(directory_path, txtname + '.txt')
    os.makedirs(directory_path, exist_ok=True)
    
    file=open(path,'w')

    for j in i['annotations']:
        label_index = label_to_idx[j['type']]
        coordinates=j['coordinates']
        coordinates_str = ' '.join(f"{coord[0]/512} {coord[1]/512}" for sublist in coordinates for coord in sublist)
        file.write(f"{label_index} {coordinates_str}\n")
```
